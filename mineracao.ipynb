## *PASSO 5:* Investigação do problema
---

Vamos po partes. Cmoeçaremos pela duração dos contratos. Ao verificar a coluna de duracao de contratos (`duracao_contrato`), varemos que existem 3 tipos de contrataos: **Anual**,**Trimensal** e **Mensal**. vejamos duracao nedia de cada tipo de contrato. Vejamos celula de codigo a sguir


a partir dessa analise, verifico que a minoria dos meus clientes possuem contrato do tipo **Mensal**. sera que a maioria dos cancelamentos e do tipo de contrato mensal? Para descobrir, vamos analisar a taxa de cancekamentos por tipo de contrato. O que vamos fazer e agrupar os 3 tipos de contrato para verificar a taxa de cancelaentos de cada um. Vejamos a celula de codigo abaixo:

```python
# numeros brutos
display(tabela['duracao_contrato'].value_counts())

# em porcentagem
display(tabela['duracao_contrato'].value_counts(normalize=True).map('{:.2%}'.format))
```

A partir dessa análise, verifico que a minoria dos meus clientes possuem contrato do tipo mensal. Será por que a Maioria dos cancelamento é do tipo de contrato mensal? Para descobrir, vamos analisar a taxa de cancelamento por tipo de contrato para verificar a taxa de cancelamentos de cada um. Vejamos a célula de código abaixo:

```python
display(tabela.groupby('duracao_contrato').mean(numeric_only=True))
```

Podemos ver que tem algo estranho  na taxa de cancelamentos vistos na coluna cancelou. Vamos verificar qual a porcentagem de cancelamentos de cada um dos tipos de contrato:

```python
tb_cancelar = tabela.groupby('duracao_contrato').mean(numeric_only=True)

display(tb_cancelar['cancelou'].map('{:.2%}'.format))
```

Conforme análise, descobrimos que os clientes que possuem contrto do tipo mensal, possuem uma taxa de 100% de cancelamento. Dessa forma, vamos re-analisar o dataset, mas desta vez, sem os contratos do tipo mensal.

```python
tabela = tabela[tabela['duracao_contrato'] != 'Monthly']
display(tabela)
display(tabela['cancelou'].value_counts())
display(tabela['cancelou'].value_counts(normalize=True).map('{:.2%}'.format))
```

conseguimos diminuir a taxa de cancelamento para **46,05%**. É uma diminuição de 10% mas a taxa de cancelamentos ainda está muito alta. Afinal, são quase 50% de cancelamentos, mesmo com a eliminação dos contratos do tipo mensal. Então vamos continuar investigando.

Pudemos verificar em análises anteriores que há uma alta taxa de cancelamento entre os cliente que atrasam o pagamento de seus contratos. Então o nosso próximo passo será analisar esses clientes.
